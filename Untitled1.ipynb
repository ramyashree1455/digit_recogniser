```python
import tensorflow as tf
from tensorflow.keras import layers, models
import pandas as pd
import numpy as np

# Load data - but you need the CSV files locally!
# Download train.csv from Kaggle competition page first
train_df = pd.read_csv('train.csv')

y_train = train_df['label'].values
X_train = train_df.drop('label', axis=1).values / 255.0
X_train = X_train.reshape(-1, 28, 28)

model = models.Sequential([
    layers.Input(shape=(28, 28)),
    layers.Flatten(),
    layers.Dense(150, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train, y_train, epochs=50, validation_split=0.1)

# Save directly into your project folder!
model.save('C:/Users/user/Desktop/new project/digit_model.keras')
print("Saved!")
```
